<h2>Grids</h2>
<p>Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.</p>

<li>We Will see below</li>
<li>PairGrid</li>
<li>Facet Grid</li>
<li>JointGrid</li></li>




```python
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print(sns.utils.get_data_home()) 
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()   # The type of flower species
```

```python
sns.pairplot(iris)
```

<h2>PairGrid</h2>
<p>Pairgrid is a subplot grid for plotting pairwise relationships in a dataset. using PairGrid we can customise pairplot</p>

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

<h2>Facet Grid</h2>
<p>FacetGrid is the general way to create grids of plots based off of a feature:</p>

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

<h2>JointGrid</h2>
<p>JointGrid is the general version for jointplot() type grids</p>

```python
tips.head()
```

```python
g=sns.JointGrid(x='total_bill',y='tip',data=tips)
g=g.plot(sns.regplot,sns.distplot)
```
